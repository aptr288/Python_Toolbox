# Formatting Dataframes

```python
zoo_eats = pd.DataFrame([['elephant','vegetables'], ['tiger','meat'], ['kangaroo','vegetables'], ['zebra','vegetables'], ['giraffe','vegetables']], columns=['animal', 'food'])
```

```python
zoo_eats
```

```python
zoodf = zdf.copy()
```

This by default is inner join so Lion and girrefe which doesnt have comman values so they were eliminated

```python
zoodf.merge(zoo_eats)
```

Here we have outter join where lion got Nan values for food and girrefe got Nan for water and uniqe columns

We have Nan where we didn’t get information from the other table.

```python
zoodf.merge(zoo_eats, how = 'outer')
```

 Keeping the giraffe line would be misleading and irrelevant since we don’t have any giraffes in our zoo anyway. That’s when merging with a how = 'left' parameter becomes handy!

```python
zoodf.merge(zoo_eats, how = 'left', left_on = 'animal', right_on = 'animal')
```

The how = 'left' parameter brought all the values from the left table (zoo) but brought only those values from the right table (zoo_eats) that we have in the left one

```python
zoodf.sort_values(by = ['animal', 'water_need'])
```

 As we can see above post operataions the index got changed. Wrong indexing can mess up your visualizations (more about that in my matplotlib tutorials) or even your machine learning models.

The point is: in certain cases, when you have done a transformation on your dataframe, you have to re-index the rows. For that, you can use the reset_index() method. 

```python
zoodf.sort_values(by = ['animal', 'water_need'], ascending = False).reset_index()
```

As you can see, our new dataframe kept the old indexes, too. If you want to remove them, just add the drop = True parameter:

```python
zoodf.sort_values(by = ['animal', 'water_need'], ascending = False).reset_index(drop = True)
```
